```python
from html.parser import HTMLParser
import pandas as pd
import re
```

```python
class MyHTMLParser(HTMLParser):
    def __init__(self, *, convert_charrefs: bool = ...) -> None:
        super().__init__(convert_charrefs=convert_charrefs)
        self.list_of_hrefs = []
        self.list_of_data = []
        self.buffer = None

    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            self.buffer = attrs[0][1]
            # self.list_of_hrefs.append(attrs[0][1])

    def handle_data(self, data):
        # self.list_of_data.append('.'.join(data.split('.')[0:-1]))
        if re.search(r'Статья [\d\.]+', data):
            self.list_of_hrefs.append('http://www.consultant.ru' + self.buffer)
            self.buffer = None
            self.list_of_data.append(re.search(r'Статья [\d\.]+', data).group()[:-1])
```

```python
f = open("text.txt", "r", encoding='utf-8')
```

```python
parser = MyHTMLParser()
parser.feed(f.read())
```

```python
zip_list = list(
    filter(lambda x: x[0] in df['article_number'].to_list(), zip(parser.list_of_data, parser.list_of_hrefs)))
list_of_data = []
list_of_hrefs = []
for row in zip_list:
    list_of_data.append(row[0])
    list_of_hrefs.append(row[1])
```

```python
df_hrefs = pd.DataFrame({
    'data': list_of_data,
    'href': list_of_hrefs
})
```

```python
# clean_df = df_hrefs[df_hrefs['data'].str.startswith('Статья')]
clean_df = df_hrefs
```

```python
clean_df.reset_index(inplace=True)
```

```python
clean_df = clean_df.drop(columns=['data', 'index'])
```

```python
clean_df
```

```python
final_data = pd.concat([df, clean_df], axis=1)
```

```python
with pd.ExcelWriter("Трудовой кодекс.xlsx", engine="openpyxl") as writer:
    final_data.to_excel(writer, 'good', engine='xlsxwriter', index=False)
    sheets_good = writer.sheets['good']
    # sheets_good.autofilter(0, 0, final_data.shape[0], final_data.shape[1])
    sheets_good.auto_filter.ref = sheets_good.dimensions
    print("Файл создан")
```
