```python
import pandas as pd
from pymongo import MongoClient
```

```python
# conn = MongoClient('192.168.10.36', 27017)
conn = MongoClient(
    'mongodb://192.168.10.36:27017/?authSource=admin&readPreference=secondary&directConnection=true&ssl=false')
gpn_conn = conn['gpn']
```

```python
gpn_conn['documents']
```

```python
%%time
document_collection = gpn_conn['documents'].find(
    {'user':{'$ne': None}, 'documentType':'CONTRACT'})
document_collection
```

```python
%%time
list_of = list(document_collection)
list_of[0]
```

```python
from bson import json_util
import json

json.loads(json_util.dumps(list_of[0], ensure_ascii=False))
```

```python
from bson import json_util
import json

for doc in list_of:
    print(doc['_id'])
    json_object = json.dumps(json.loads(json_util.dumps(doc)), indent=4, ensure_ascii=False)
    with open(f"./data/{doc['_id']}.json", "w", encoding='utf-8') as outfile:
        outfile.write(json_object)
```
